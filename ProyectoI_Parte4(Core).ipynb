- Crea nuevas columnas: Basándonos en los datos existentes, crea nuevas columnas que sean útiles para el análisis. Por ejemplo, calcula el ingreso total por venta y normaliza las ventas.

```python
max = df['Price per Unit'].max()
min = df['Price per Unit'].min()
# Preprocesamiento de datos para escalar valores en un rango específico.
df['Normalización'] = df['Price per Unit'].apply(lambda x: (x - min)/ (max - min)).round(2)
df.head()
```

```python
# Descripción de los estadísticos con la nueva columna
df.describe().round(2)
```

- Clasifica los datos: Crea una columna que clasifique las ventas en categorías significativas (e.g., ‘Alta’, ‘Media’, ‘Baja’).

```python
# Clasificación de valores
df['Clasificación ítems vendidos'] = df['Price per Unit'].apply(lambda x: 'Baja' if x <= 1 else ('Media' if x > 2 and x < 4 else 'Alta'))
df.head()
```

- Aplicar funciones de agregación: Utiliza funciones como sum, mean, count, min, max, std, y var para obtener estadísticas descriptivas de cada grupo.

## 3. Análisis Personalizado con apply

- Función personalizada: Aplica funciones personalizadas para realizar análisis específicos que no se pueden lograr con las funciones de agregación estándar.

```python
# Cálculo de la desviación absoluta media
Promedio_totalamount = df['Total Amount'].mean()
df['Desviación abosluta media'] = df['Total Amount'].apply(lambda x: (x- Promedio_totalamount))
df.head()
```

- Utiliza histogramas y boxplots para visualizar la distribución de las variables numéricas clave.

```python
# Crear histograma de la categoria Total Amount, Quantity y Price per unit

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

fig, axs = plt.subplots(2,3, figsize=(20,10))

sns.histplot(df['Total Amount'], kde= True, color = 'skyblue', bins= 4, edgecolor= 'black', alpha= 0.3, ax=axs[0,0])
sns.boxplot(df['Total Amount'], notch= True, patch_artist = True,
            boxprops = dict(facecolor = 'black', color = 'skyblue'),
            medianprops= dict(color= 'r'), whiskerprops= dict(color='black'), capprops = dict(color='black'), ax=axs[1,0])
sns.histplot(df['Quantity'], kde= True, color = 'powderblue', bins= 4, edgecolor= 'black', alpha= 0.3, ax=axs[0,1])
sns.boxplot(df['Quantity'], notch= True, patch_artist = True,
            boxprops = dict(facecolor = 'black', color = 'powderblue'),
            medianprops= dict(color= 'r'), whiskerprops= dict(color='black'), capprops = dict(color='black'), ax=axs[1,1])
sns.histplot(df['Price per Unit'], kde= True, color = 'paleturquoise', bins= 4, edgecolor= 'black', alpha= 0.3, ax=axs[0,2])
sns.boxplot(df['Price per Unit'], notch= True, patch_artist = True,
            boxprops = dict(facecolor = 'black', color = 'paleturquoise'),
            medianprops= dict(color= 'r'), whiskerprops= dict(color='black'), capprops = dict(color='black'), ax=axs[1,2])
plt.tight_layout(pad=2.0, h_pad=2.0, w_pad=2.0)
plt.show()

```

-Gráfico 1 Total Amount (histograma). hay una tendencia asimétrica lo cual tiene sentido ya que es coherente que se desarrolle un crecimiento de ventas, se puede identifcar también aquellas ventas desde 750 hasta 2000 con baja frecuencia.

-Gráfico Total Amount (boxplot). Indica que la media de pagos en ventas está casi en 250, que hay presencia de un dato atípico que puede ser producto de alguna novedad, hay un flujo de ventas mayormente concentradas en la parte superior del gráfico.

-Gráfico 2 Quantity (histograma). tendencia multimodal en la que se puede identifcar ventas de los artículos en cantidades muy similares.

-Gráfico Quantity (boxplot). Indica que la media de artículo por venta es de 3 unidades, no se evidencia presencia de valores atípicos.

-Gráfico 3 Price per unit (histograma). tendencia multimodal en la que se puede apreciar al igual que es Quantity ventas en proporciones muy similares con una cantidad máxima por producto de 4.0.

-Gráfico Price per unit (boxplot). Al igual que en el gráfico de Quantity una media de 3.0 que corresponde al precio del producto, sin presencia de datos atípicos, mayor concentración de los datos en la parte inferior indica que el rango precios en el que más se vende.

- Combina histogramas y boxplots para visualizar la distribución de las ventas y detectar valores atípicos.

```python
genero = df.groupby(['Gender'])['Total Amount'].sum()
genero
```

```python
#Crear gráfico con las categorias genero para visualizar el comportamiento de compra por género
X = plt.figure(figsize=(20, 10))

# Gráfico de barras para la frecuencia de compras por género
XX = X.add_subplot(1, 2, 1)
genero.plot(kind='bar', color=['powderblue', 'cadetblue'], edgecolor='black', alpha=0.7, ax=XX)
XX.set_title('Frecuencia de compra por género')
XX.set_xlabel('Género')
XX.set_ylabel('Frecuencia')

# Boxplot
XX1 = X.add_subplot(1, 2, 2)
# Preparar datos para el boxplot
data_for_boxplot = [df[df['Gender'] == gender]['Total Amount'] for gender in genero.index]
XX1.boxplot(data_for_boxplot, labels=genero.index, notch=True, patch_artist=True,
            boxprops=dict(facecolor='powderblue', color='black'), medianprops=dict(color='red'))
XX1.set_title('Ventas por género')
XX1.grid()

plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

Gráfico de barras en dónde se evidencia mayor frecuencia de compra por parte de las mujeres, en el boxplots se evidencia un comportamiento similar para ambos géneros.
